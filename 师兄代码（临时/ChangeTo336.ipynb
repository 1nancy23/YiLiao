### 图像得到特征

```python

import torch
import torchvision.models as models
from torchvision import transforms
from PIL import Image
import os
import numpy as np
from pathlib import Path

# 加载预训练的 MobileNetV3 模型
model = models.mobilenet_v3_large(pretrained=True)
model = torch.nn.Sequential(*list(model.children())[:-1])
model.eval()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# 图像预处理
preprocess = transforms.Compose([
    transforms.Resize(224),
    # transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

def extract_and_save_features(input_folder, output_folder="features"):
    """批量提取并单独保存每张图片的特征向量"""
    # 创建输出文件夹
    Path(output_folder).mkdir(exist_ok=True)
    
    # 遍历所有图片
    image_files = [f for f in os.listdir(input_folder) 
                   if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    for filename in image_files:
        # 读取图片
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path).convert('RGB')
        img_tensor = preprocess(img).unsqueeze(0).to(device)
        
        # 提取特征
        with torch.no_grad():
            features = model(img_tensor)
            features = features.squeeze().cpu().numpy()
        
        # 保存特征向量
        base_name = os.path.splitext(filename)[0]
        output_path = os.path.join(output_folder, f"{base_name}.npy")
        np.save(output_path, features)
        
        print(f"已处理: {filename} -> {base_name}.npy (维度: {features.shape})")
    
    print(f"\n完成！共处理 {len(image_files)} 张图片，特征已保存到 {output_folder} 文件夹")

# 使用
if __name__ == "__main__":
    for i in range(1,8):
        input_folder = f"A:\Temp_dataset\Blured\\Blured_resized_{i}"  # 当前文件夹，可修改为你的图片路径
        output_folder = f"A:\Temp_dataset\Blured\\features_resized_{i}"

    
        extract_and_save_features(input_folder, output_folder)
```

### 单个相似度测试

```python
import numpy as np
import os
from pathlib import Path

def load_features_from_folders(base_folders):
    """从多个文件夹加载特征向量"""
    all_features = []
    all_labels = []
    all_filenames = []
    
    for folder_idx, folder_path in enumerate(base_folders):
        npy_files = [f for f in os.listdir(folder_path) if f.endswith('.npy')]
        
        for npy_file in npy_files:
            feature_path = os.path.join(folder_path, npy_file)
            feature = np.load(feature_path)
            
            all_features.append(feature)
            all_labels.append(folder_idx)  # 文件夹索引作为类别标签
            all_filenames.append((folder_path, npy_file))
        
        print(f"已加载文件夹 {folder_path}: {len(npy_files)} 个特征")
    
    return np.array(all_features), np.array(all_labels), all_filenames

def cosine_similarity(vec1, vec2):
    """计算余弦相似度"""
    dot_product = np.dot(vec1, vec2)
    norm1 = np.linalg.norm(vec1)
    norm2 = np.linalg.norm(vec2)
    return dot_product / (norm1 * norm2)

def euclidean_distance(vec1, vec2):
    """计算欧氏距离（越小越相似）"""
    return np.linalg.norm(vec1 - vec2)

def find_most_similar(query_feature, all_features, all_labels, all_filenames, 
                      top_k=5, metric='cosine'):
    """找到最相似的特征向量"""
    similarities = []
    
    for idx, feature in enumerate(all_features):
        if metric == 'cosine':
            sim = cosine_similarity(query_feature, feature)
            similarities.append((sim, idx))
        elif metric == 'euclidean':
            dist = euclidean_distance(query_feature, feature)
            similarities.append((-dist, idx))  # 负号使得距离小的排在前面
    
    # 排序（相似度从高到低）
    similarities.sort(reverse=True, key=lambda x: x[0])
    
    # 返回 top_k 个结果
    results = []
    for i in range(min(top_k, len(similarities))):
        sim_score, idx = similarities[i]
        results.append({
            'similarity': sim_score if metric == 'cosine' else -sim_score,
            'label': all_labels[idx],
            'folder': all_filenames[idx][0],
            'filename': all_filenames[idx][1]
        })
    
    return results

def predict_category(query_feature, all_features, all_labels, metric='cosine', k=5):
    """基于 k 近邻预测类别"""
    similarities = []
    
    for idx, feature in enumerate(all_features):
        if metric == 'cosine':
            sim = cosine_similarity(query_feature, feature)
        elif metric == 'euclidean':
            sim = -euclidean_distance(query_feature, feature)
        similarities.append((sim, all_labels[idx]))
    
    # 排序并取前 k 个
    similarities.sort(reverse=True, key=lambda x: x[0])
    top_k_labels = [label for _, label in similarities[:k]]
    
    # 投票决定类别
    predicted_label = max(set(top_k_labels), key=top_k_labels.count)
    confidence = top_k_labels.count(predicted_label) / k
    
    return predicted_label, confidence

# 使用示例
if __name__ == "__main__":
    # 定义五个文件夹路径
    folders = [
        "A:\Temp_dataset\Blured\\features_resized_1",
        "A:\Temp_dataset\Blured\\features_resized_2",
        "A:\Temp_dataset\Blured\\features_resized_3",
        "A:\Temp_dataset\Blured\\features_resized_4",
        "A:\Temp_dataset\Blured\\features_resized_5",
        "A:\Temp_dataset\Blured\\features_resized_6",
        "A:\Temp_dataset\Blured\\features_resized_7"
    ]
    
    # 加载所有特征
    print("正在加载所有特征向量...")
    all_features, all_labels, all_filenames = load_features_from_folders(folders)
    print(f"\n总共加载 {len(all_features)} 个特征向量\n")
    
    # 加载待匹配的特征向量
    query_path = "A:\项目\医药物品检测\DataBase_Data\YaoPing\YaoPing_test_true\Features_Cropped_Resized\\frame_000783.npy"  # 待匹配的特征文件
    query_feature = np.load(query_path)
    
    # 方法1: 找到最相似的 top-k 个特征
    print("=" * 60)
    print("方法1: 找到最相似的特征")
    print("=" * 60)
    top_results = find_most_similar(query_feature, all_features, all_labels, 
                                     all_filenames, top_k=3, metric='cosine')
    
    for i, result in enumerate(top_results, 1):
        print(f"Top {i}:")
        print(f"  相似度: {result['similarity']:.4f}")
        print(f"  类别: {result['label']}")
        print(f"  文件夹: {result['folder']}")
        print(f"  文件名: {result['filename']}")
        print()
    
    # 方法2: 基于 k 近邻预测类别
    print("=" * 60)
    print("方法2: K近邻分类预测")
    print("=" * 60)
    predicted_label, confidence = predict_category(query_feature, all_features, 
                                                    all_labels, metric='cosine', k=3)
    
    print(f"预测类别: {predicted_label}")
    print(f"置信度: {confidence:.2%}")
    print(f"对应文件夹: {folders[predicted_label]}")
```

### 高斯模糊图片

```python
"""
批量图像模糊处理 - 模拟真实失焦效果
"""

import os
import cv2
import numpy as np
from pathlib import Path
from tqdm import tqdm


def apply_defocus_blur(image, blur_level):
    """
    应用失焦模糊效果（更接近真实相机失焦）
    
    参数:
        image: 输入图像
        blur_level: 模糊程度 (0-100)
    
    返回:
        模糊后的图像
    """
    if blur_level == 0:
        return image
    
    # 将模糊程度映射到合适的范围
    # 对于轻微模糊，使用较小的核
    kernel_size = max(3, int(blur_level * 0.15) + 1)
    if kernel_size % 2 == 0:
        kernel_size += 1
    
    # 使用双边滤波 + 轻微高斯模糊组合，更接近真实失焦
    # 先用双边滤波保留边缘
    bilateral = cv2.bilateralFilter(image, 5, 50, 50)
    
    # 再应用轻微高斯模糊
    sigma = blur_level * 0.05  # 较小的sigma值
    blurred = cv2.GaussianBlur(bilateral, (kernel_size, kernel_size), sigma)
    
    # 混合原图和模糊图，产生更自然的效果
    alpha = min(0.7, blur_level / 150.0)  # 混合比例
    result = cv2.addWeighted(image, 1 - alpha, blurred, alpha, 0)
    
    return result


def apply_motion_blur(image, blur_level):
    """
    应用运动模糊效果
    
    参数:
        image: 输入图像
        blur_level: 模糊程度 (0-100)
    
    返回:
        模糊后的图像
    """
    if blur_level == 0:
        return image
    
    # 创建运动模糊核
    kernel_size = max(3, int(blur_level * 0.2) + 1)
    kernel = np.zeros((kernel_size, kernel_size))
    kernel[int((kernel_size - 1) / 2), :] = np.ones(kernel_size)
    kernel = kernel / kernel_size
    
    # 应用运动模糊
    blurred = cv2.filter2D(image, -1, kernel)
    
    return blurred


def apply_lens_blur(image, blur_level):
    """
    应用镜头模糊效果（圆形散焦）
    
    参数:
        image: 输入图像
        blur_level: 模糊程度 (0-100)
    
    返回:
        模糊后的图像
    """
    if blur_level == 0:
        return image
    
    # 创建圆形核
    kernel_size = max(3, int(blur_level * 0.15) + 1)
    if kernel_size % 2 == 0:
        kernel_size += 1
    
    kernel = np.zeros((kernel_size, kernel_size))
    center = kernel_size // 2
    radius = center
    
    y, x = np.ogrid[-center:center+1, -center:center+1]
    mask = x**2 + y**2 <= radius**2
    kernel[mask] = 1
    kernel = kernel / np.sum(kernel)
    
    # 应用模糊
    blurred = cv2.filter2D(image, -1, kernel)
    
    return blurred


def apply_realistic_blur(image, blur_level, blur_type='defocus'):
    """
    应用真实模糊效果
    
    参数:
        image: 输入图像
        blur_level: 模糊程度 (0-100)
        blur_type: 模糊类型 ('defocus', 'motion', 'lens', 'gaussian')
    
    返回:
        模糊后的图像
    """
    if blur_type == 'defocus':
        return apply_defocus_blur(image, blur_level)
    elif blur_type == 'motion':
        return apply_motion_blur(image, blur_level)
    elif blur_type == 'lens':
        return apply_lens_blur(image, blur_level)
    else:  # gaussian
        if blur_level == 0:
            return image
        kernel_size = int(blur_level * 0.98) + 1
        if kernel_size % 2 == 0:
            kernel_size += 1
        return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)


def batch_blur_images(
    input_folder,
    output_folder,
    blur_level=50,
    blur_type='defocus',
    supported_formats=None
):
    """
    批量对文件夹中的图片应用模糊
    
    参数:
        input_folder: 输入文件夹路径
        output_folder: 输出文件夹路径
        blur_level: 模糊程度 (0-100)
        blur_type: 模糊类型 ('defocus', 'motion', 'lens', 'gaussian')
        supported_formats: 支持的图片格式列表
    
    返回:
        处理成功的文件数量
    """
    
    if supported_formats is None:
        supported_formats = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif']
    
    os.makedirs(output_folder, exist_ok=True)
    
    image_files = []
    for ext in supported_formats:
        image_files.extend(Path(input_folder).glob(f'*{ext}'))
        image_files.extend(Path(input_folder).glob(f'*{ext.upper()}'))
    
    if not image_files:
        print(f"❌ 在 {input_folder} 中未找到图片文件")
        return 0
    
    print(f"\n{'='*60}")
    print(f"批量模糊处理")
    print(f"{'='*60}")
    print(f"输入文件夹: {input_folder}")
    print(f"输出文件夹: {output_folder}")
    print(f"模糊类型: {blur_type}")
    print(f"模糊程度: {blur_level}/100")
    print(f"找到图片: {len(image_files)} 张")
    print(f"{'='*60}\n")
    
    success_count = 0
    failed_files = []
    
    for img_path in tqdm(image_files, desc="处理进度"):
        try:
            image = cv2.imread(str(img_path))
            
            if image is None:
                print(f"⚠️ 无法读取: {img_path.name}")
                failed_files.append(img_path.name)
                continue
            
            blurred_image = apply_realistic_blur(image, blur_level, blur_type)
            
            output_path = os.path.join(output_folder, img_path.name)
            cv2.imwrite(output_path, blurred_image)
            
            success_count += 1
            
        except Exception as e:
            print(f"❌ 处理失败 {img_path.name}: {str(e)}")
            failed_files.append(img_path.name)
    
    print(f"\n{'='*60}")
    print(f"处理完成")
    print(f"{'='*60}")
    print(f"成功: {success_count}/{len(image_files)}")
    print(f"失败: {len(failed_files)}")
    
    if failed_files:
        print(f"\n失败的文件:")
        for f in failed_files:
            print(f"  - {f}")
    
    print(f"{'='*60}\n")
    
    return success_count


def create_blur_comparison(
    image_path,
    output_path,
    blur_levels=None,
    blur_type='defocus'
):
    """
    创建单张图片的模糊对比图
    """
    
    if blur_levels is None:
        blur_levels = [0, 10, 20, 30, 40, 50]
    
    image = cv2.imread(image_path)
    if image is None:
        print(f"❌ 无法读取图片: {image_path}")
        return
    
    n_images = len(blur_levels)
    cols = 3
    rows = (n_images + cols - 1) // cols
    
    sub_h, sub_w = 224, 224
    canvas = np.ones((rows * sub_h + rows * 40, cols * sub_w, 3), dtype=np.uint8) * 255
    
    for idx, level in enumerate(blur_levels):
        row = idx // cols
        col = idx % cols
        
        blurred = apply_realistic_blur(image, level, blur_type)
        resized = cv2.resize(blurred, (sub_w, sub_h))
        
        y_start = row * (sub_h + 40)
        x_start = col * sub_w
        canvas[y_start:y_start+sub_h, x_start:x_start+sub_w] = resized
        
        label = f"{blur_type.capitalize()}: {level}"
        cv2.putText(canvas, label, (x_start + 10, y_start + sub_h + 30),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2)
    
    cv2.imwrite(output_path, canvas)
    print(f"✅ 对比图已保存: {output_path}")


# ==================== 主程序 ====================

if __name__ == "__main__":
    
    # 根据你的图片，推荐使用 defocus 类型，blur_level 在 15-30 之间
    
    print("\n轻微失焦模糊处理（类似你的示例图）")
    for i in range(1,14):
        batch_blur_images(
            input_folder=f"A:\\Temp_dataset\\resized_{i}",
            output_folder=f"A:\\Temp_dataset\\Blured\\Blured_resized_{i}",
            blur_level=50,           # 轻微模糊，推荐 15-30
            blur_type='defocus'      # 失焦效果
        )
        
        # 如果想测试不同效果，可以取消下面的注释
        """
        # 测试不同模糊类型
        for blur_type in ['defocus', 'lens', 'gaussian']:
            print(f"\n测试 {blur_type} 模糊")
            batch_blur_images(
                input_folder=r"A:\Temp_dataset\resized_7",
                output_folder=rf"A:\Temp_dataset\Blured\{blur_type}_20",
                blur_level=20,
                blur_type=blur_type
            )
        """
```

# CV算法直接匹配

```python
import cv2
import numpy as np
import torch
import torchvision.models as models
from torchvision import transforms
from PIL import Image
import os
from pathlib import Path
from scipy.spatial.distance import cosine
import json
from tqdm import tqdm

class PharmaceuticalBottleClassifier:
    """药瓶多类别分类器"""
    
    def __init__(self, device='cuda'):
        self.device = torch.device(device if torch.cuda.is_available() else 'cpu')
        
        # SIFT 特征提取
        self.sift = cv2.SIFT_create()
        
        # 深度学习特征提取
        self.model = models.mobilenet_v3_large(pretrained=True)
        self.model = torch.nn.Sequential(*list(self.model.children())[:-1])
        self.model.eval()
        self.model = self.model.to(self.device)
        
        self.preprocess = transforms.Compose([
            transforms.Resize(224),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                               std=[0.229, 0.224, 0.225]),
        ])
        
        # 特征匹配器
        FLANN_INDEX_KDTREE = 1
        index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
        search_params = dict(checks=50)
        self.flann = cv2.FlannBasedMatcher(index_params, search_params)
        
        # 存储每个类别的模板特征
        self.category_templates = {}  # {category_name: [features_dict, ...]}
    
    def preprocess_image(self, image):
        """预处理图像"""
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
        enhanced = clahe.apply(gray)
        
        kernel = np.array([[-1, -1, -1],
                          [-1,  9, -1],
                          [-1, -1, -1]], dtype=np.float32)
        sharpened = cv2.filter2D(enhanced, -1, kernel)
        
        return sharpened
    
    def extract_sift_features(self, image):
        """提取 SIFT 特征"""
        processed = self.preprocess_image(image)
        kp, desc = self.sift.detectAndCompute(processed, None)
        return kp, desc
    
    def extract_deep_features(self, image):
        """提取深度学习特征"""
        img_pil = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        img_tensor = self.preprocess(img_pil).unsqueeze(0).to(self.device)
        
        with torch.no_grad():
            features = self.model(img_tensor)
            features = features.squeeze().cpu().numpy()
        
        return features
    
    def match_sift_features(self, desc1, desc2):
        """SIFT 特征匹配"""
        if desc1 is None or desc2 is None:
            return []
        
        matches = self.flann.knnMatch(desc1, desc2, k=2)
        good_matches = []
        
        for match_pair in matches:
            if len(match_pair) == 2:
                m, n = match_pair
                if m.distance < 0.7 * n.distance:
                    good_matches.append(m)
        
        return good_matches
    
    def match_with_rotation(self, template_img, query_img, 
                           rotation_range=(-45, 45), step=5):
        """处理旋转的匹配"""
        best_matches = []
        best_angle = 0
        best_score = 0
        
        kp_template, desc_template = self.extract_sift_features(template_img)
        
        h, w = query_img.shape[:2]
        center = (w // 2, h // 2)
        
        for angle in range(rotation_range[0], rotation_range[1] + 1, step):
            M = cv2.getRotationMatrix2D(center, angle, 1.0)
            rotated = cv2.warpAffine(query_img, M, (w, h))
            
            kp_query, desc_query = self.extract_sift_features(rotated)
            matches = self.match_sift_features(desc_template, desc_query)
            
            if len(matches) > best_score:
                best_score = len(matches)
                best_matches = matches
                best_angle = angle
        
        return best_matches, best_angle, best_score
    
    def load_category_templates(self, base_folder):
        """
        加载所有类别的模板图像
        
        参数:
            base_folder: 基础文件夹路径，其中包含多个类别子文件夹
                        结构: base_folder/
                             ├── category_1/
                             │   ├── template_1.jpg
                             │   ├── template_2.jpg
                             │   └── ...
                             ├── category_2/
                             │   ├── template_1.jpg
                             │   └── ...
                             └── ...
        """
        print(f"正在加载模板图像...")
        
        category_folders = [d for d in os.listdir(base_folder) 
                           if os.path.isdir(os.path.join(base_folder, d))]
        
        for category in category_folders:
            category_path = os.path.join(base_folder, category)
            image_files = [f for f in os.listdir(category_path) 
                          if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
            
            self.category_templates[category] = []
            
            for img_file in image_files:
                img_path = os.path.join(category_path, img_file)
                img = cv2.imread(img_path)
                
                if img is None:
                    continue
                
                # 提取特征
                kp_sift, desc_sift = self.extract_sift_features(img)
                deep_feat = self.extract_deep_features(img)
                
                template_info = {
                    'filename': img_file,
                    'path': img_path,
                    'kp_sift': kp_sift,
                    'desc_sift': desc_sift,
                    'deep_features': deep_feat
                }
                
                self.category_templates[category].append(template_info)
            
            print(f"  类别 '{category}': 加载 {len(image_files)} 张模板")
        
        print(f"总共加载 {len(category_folders)} 个类别\n")
    
    def classify_single_image(self, query_img, use_rotation_correction=True):
        """
        对单个查询图像进行分类
        
        返回:
            分类结果字典
        """
        # 提取查询图像特征
        kp_query, desc_query = self.extract_sift_features(query_img)
        deep_feat_query = self.extract_deep_features(query_img)
        
        category_scores = {}
        category_details = {}
        
        # 对每个类别进行匹配
        for category, templates in self.category_templates.items():
            sift_scores = []
            deep_scores = []
            rotation_angles = []
            
            for template in templates:
                # SIFT 匹配
                if use_rotation_correction:
                    matches, angle, score = self.match_with_rotation(
                        cv2.imread(template['path']), query_img
                    )
                    rotation_angles.append(angle)
                else:
                    matches = self.match_sift_features(
                        template['desc_sift'], desc_query
                    )
                    score = len(matches)
                    rotation_angles.append(0)
                
                sift_scores.append(score)
                
                # 深度学习特征相似度
                deep_sim = 1 - cosine(template['deep_features'], deep_feat_query)
                deep_scores.append(deep_sim)
            
            # 计算该类别的综合得分
            avg_sift_score = np.mean(sift_scores) if sift_scores else 0
            avg_deep_score = np.mean(deep_scores) if deep_scores else 0
            max_sift_score = np.max(sift_scores) if sift_scores else 0
            max_deep_score = np.max(deep_scores) if deep_scores else 0
            
            # 归一化
            sift_confidence = min(1.0, max_sift_score / 15.0)
            deep_confidence = max_deep_score
            
            # 加权融合
            final_score = sift_confidence * 0.6 + deep_confidence * 0.4
            
            category_scores[category] = final_score
            category_details[category] = {
                'sift_scores': sift_scores,
                'deep_scores': deep_scores,
                'avg_sift_score': avg_sift_score,
                'avg_deep_score': avg_deep_score,
                'max_sift_score': max_sift_score,
                'max_deep_score': max_deep_score,
                'sift_confidence': sift_confidence,
                'deep_confidence': deep_confidence,
                'rotation_angles': rotation_angles
            }
        
        # 排序得分
        sorted_categories = sorted(category_scores.items(), 
                                  key=lambda x: x[1], reverse=True)
        
        # 获取最佳匹配
        best_category = sorted_categories[0][0]
        best_score = sorted_categories[0][1]
        
        result = {
            'predicted_category': best_category,
            'confidence': best_score,
            'all_scores': dict(sorted_categories),
            'details': category_details,
            'top_3': sorted_categories[:3]
        }
        
        return result
    
    def classify_batch(self, query_folder, output_json=None, output_csv=None):
        """
        批量分类查询图像
        
        参数:
            query_folder: 查询图像文件夹
            output_json: 输出 JSON 结果文件
            output_csv: 输出 CSV 结果文件
        """
        query_files = [f for f in os.listdir(query_folder) 
                      if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
        
        results = []
        
        print(f"正在分类 {len(query_files)} 张查询图像...\n")
        
        for query_file in tqdm(query_files, desc="分类进度"):
            query_path = os.path.join(query_folder, query_file)
            query_img = cv2.imread(query_path)
            
            if query_img is None:
                continue
            
            result = self.classify_single_image(query_img)
            result['query_file'] = query_file
            results.append(result)
        
        # 保存结果
        if output_json:
            self._save_json_results(results, output_json)
        
        if output_csv:
            self._save_csv_results(results, output_csv)
        
        return results
    
    def _save_json_results(self, results, output_path):
        """保存为 JSON"""
        json_results = []
        
        for result in results:
            json_result = {
                'query_file': result['query_file'],
                'predicted_category': result['predicted_category'],
                'confidence': float(result['confidence']),
                'all_scores': {k: float(v) for k, v in result['all_scores'].items()},
                'top_3': [(cat, float(score)) for cat, score in result['top_3']]
            }
            json_results.append(json_result)
        
        with open(output_path, 'w', encoding='utf-8') as f:
            json.dump(json_results, f, ensure_ascii=False, indent=2)
        
        print(f"JSON 结果已保存到: {output_path}")
    
    def _save_csv_results(self, results, output_path):
        """保存为 CSV"""
        import csv
        
        with open(output_path, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(['查询文件', '预测类别', '置信度', '所有类别得分'])
            
            for result in results:
                scores_str = '; '.join([f"{k}:{v:.4f}" 
                                       for k, v in result['all_scores'].items()])
                writer.writerow([
                    result['query_file'],
                    result['predicted_category'],
                    f"{result['confidence']:.4f}",
                    scores_str
                ])
        
        print(f"CSV 结果已保存到: {output_path}")
    
    def print_classification_report(self, result):
        """打印分类报告"""
        print("\n" + "="*70)
        print("分类结果")
        print("="*70)
        print(f"查询文件: {result['query_file']}")
        print(f"预测类别: {result['predicted_category']}")
        print(f"置信度: {result['confidence']:.4f}")
        print("\n所有类别得分:")
        
        for category, score in result['all_scores'].items():
            print(f"  {category}: {score:.4f}")
        
        print("\nTop 3 预测:")
        for i, (category, score) in enumerate(result['top_3'], 1):
            details = result['details'][category]
            print(f"  {i}. {category} (置信度: {score:.4f})")
            print(f"     - SIFT 最高分: {details['max_sift_score']:.1f}")
            print(f"     - 深度学习最高分: {details['max_deep_score']:.4f}")
            print(f"     - 旋转角度: {details['rotation_angles']}")
        
        print("="*70 + "\n")


# ==================== 使用示例 ====================

if __name__ == "__main__":
    
    # 初始化分类器
    classifier = PharmaceuticalBottleClassifier()
    
    # 加载模板（多个类别文件夹）
    base_template_folder = r"A:\\Temp_dataset\\Blured"
    # 结构应该是:
    # YaoPing/
    # ├── 1/
    # │   ├── img1.jpg
    # │   ├── img2.jpg
    # │   └── ...
    # ├── 2/
    # │   ├── img1.jpg
    # │   └── ...
    # └── ...
    
    classifier.load_category_templates(base_template_folder)
    
    # ========== 方式1: 单个图像分类 ==========
    print("\n方式1: 单个图像分类")
    print("-" * 70)
    
    query_image_path = "./YaoPing_test_true/Cropped_Resized/frame_000293_2.png"
    query_img = cv2.imread(query_image_path)
    
    result = classifier.classify_single_image(query_img, use_rotation_correction=True)
    result['query_file'] = os.path.basename(query_image_path)
    
    classifier.print_classification_report(result)
    
    # ========== 方式2: 批量分类 ==========
    # print("\n方式2: 批量分类")
    # print("-" * 70)
    
    # query_folder = r"A:\项目\医药物品检测\DataBase_Data\YaoPing\YaoPing_test_true"
    
    # batch_results = classifier.classify_batch(
    #     query_folder,
    #     output_json=r"A:\项目\医药物品检测\classification_results.json",
    #     output_csv=r"A:\项目\医药物品检测\classification_results.csv"
    # )
    
    # # 打印前几个结果
    # for result in batch_results[:3]:
    #     classifier.print_classification_report(result)
    
    # # 统计准确率（如果有真实标签）
    # print("\n分类统计:")
    # print("-" * 70)
    # category_counts = {}
    # for result in batch_results:
    #     cat = result['predicted_category']
    #     category_counts[cat] = category_counts.get(cat, 0) + 1
    
    # for category, count in sorted(category_counts.items()):
    #     print(f"  {category}: {count} 张")
```

# 带缓存的识别分类----特征生成

```python
import cv2
import numpy as np
import torch
import torchvision.models as models
from torchvision import transforms
from PIL import Image
import os
from pathlib import Path
from scipy.spatial.distance import cosine
import json
from tqdm import tqdm
import pickle

class PharmaceuticalBottleClassifier:
    """药瓶多类别分类器"""

    def __init__(self, device='cuda', cache_dir='./feature_cache'):
        self.device = torch.device(device if torch.cuda.is_available() else 'cpu')
        self.cache_dir = cache_dir
        
        # 创建缓存目录
        os.makedirs(cache_dir, exist_ok=True)
        
        # SIFT 特征提取
        self.sift = cv2.SIFT_create()
        
        # 深度学习特征提取
        self.model = models.mobilenet_v3_large(pretrained=True)
        self.model = torch.nn.Sequential(*list(self.model.children())[:-1])
        self.model.eval()
        self.model = self.model.to(self.device)
        
        self.preprocess = transforms.Compose([
            transforms.Resize(224),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                               std=[0.229, 0.224, 0.225]),
        ])
        
        # 特征匹配器
        FLANN_INDEX_KDTREE = 1
        index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
        search_params = dict(checks=50)
        self.flann = cv2.FlannBasedMatcher(index_params, search_params)
        
        # 存储每个类别的模板特征
        self.category_templates = {}

    def _get_cache_path(self, base_folder):
        """获取缓存文件路径"""
        folder_hash = str(hash(os.path.abspath(base_folder)))
        cache_file = os.path.join(self.cache_dir, f'features_{folder_hash}.pkl')
        return cache_file

    def _save_features_cache(self, base_folder):
        """保存特征到本地缓存"""
        cache_path = self._get_cache_path(base_folder)
        
        # 准备可序列化的数据（SIFT keypoints 需要特殊处理）
        cache_data = {}
        
        for category, templates in self.category_templates.items():
            cache_data[category] = []
            
            for template in templates:
                # 转换 SIFT keypoints 为可序列化格式
                kp_data = []
                if template['kp_sift']:
                    for kp in template['kp_sift']:
                        kp_data.append({
                            'pt': kp.pt,
                            'size': kp.size,
                            'angle': kp.angle,
                            'response': kp.response,
                            'octave': kp.octave,
                            'class_id': kp.class_id
                        })
                
                cache_item = {
                    'filename': template['filename'],
                    'path': template['path'],
                    'kp_sift': kp_data,
                    'desc_sift': template['desc_sift'],
                    'deep_features': template['deep_features']
                }
                cache_data[category].append(cache_item)
        
        with open(cache_path, 'wb') as f:
            pickle.dump(cache_data, f)
        
        print(f"特征已保存到: {cache_path}\n")

    def _load_features_cache(self, base_folder):
        """从缓存加载特征"""
        cache_path = self._get_cache_path(base_folder)
        
        if not os.path.exists(cache_path):
            return False
        
        try:
            with open(cache_path, 'rb') as f:
                cache_data = pickle.load(f)
            
            # 恢复 SIFT keypoints
            for category, templates in cache_data.items():
                self.category_templates[category] = []
                
                for template in templates:
                    # 重建 keypoints
                    kp_list = []
                    for kp_data in template['kp_sift']:
                        kp = cv2.KeyPoint(
                            x=kp_data['pt'][0],
                            y=kp_data['pt'][1],
                            size=kp_data['size'],
                            angle=kp_data['angle'],
                            response=kp_data['response'],
                            octave=kp_data['octave'],
                            class_id=kp_data['class_id']
                        )
                        kp_list.append(kp)
                    
                    template_info = {
                        'filename': template['filename'],
                        'path': template['path'],
                        'kp_sift': kp_list,
                        'desc_sift': template['desc_sift'],
                        'deep_features': template['deep_features']
                    }
                    self.category_templates[category].append(template_info)
            
            print(f"从缓存加载特征: {cache_path}")
            print(f"加载了 {len(self.category_templates)} 个类别\n")
            return True
        
        except Exception as e:
            print(f"加载缓存失败: {e}")
            return False

    def preprocess_image(self, image):
        """预处理图像"""
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
        enhanced = clahe.apply(gray)
        
        kernel = np.array([[-1, -1, -1],
                          [-1,  9, -1],
                          [-1, -1, -1]], dtype=np.float32)
        sharpened = cv2.filter2D(enhanced, -1, kernel)
        
        return sharpened

    def extract_sift_features(self, image):
        """提取 SIFT 特征"""
        processed = self.preprocess_image(image)
        kp, desc = self.sift.detectAndCompute(processed, None)
        return kp, desc

    def extract_deep_features(self, image):
        """提取深度学习特征"""
        img_pil = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        img_tensor = self.preprocess(img_pil).unsqueeze(0).to(self.device)
        
        with torch.no_grad():
            features = self.model(img_tensor)
            features = features.squeeze().cpu().numpy()
        
        return features

    def match_sift_features(self, desc1, desc2):
        """SIFT 特征匹配"""
        if desc1 is None or desc2 is None:
            return []
        
        matches = self.flann.knnMatch(desc1, desc2, k=2)
        good_matches = []
        
        for match_pair in matches:
            if len(match_pair) == 2:
                m, n = match_pair
                if m.distance < 0.7 * n.distance:
                    good_matches.append(m)
        
        return good_matches

    def match_with_rotation(self, template_img, query_img, 
                           rotation_range=(-45, 45), step=5):
        """处理旋转的匹配"""
        best_matches = []
        best_angle = 0
        best_score = 0
        
        kp_template, desc_template = self.extract_sift_features(template_img)
        
        h, w = query_img.shape[:2]
        center = (w // 2, h // 2)
        
        for angle in range(rotation_range[0], rotation_range[1] + 1, step):
            M = cv2.getRotationMatrix2D(center, angle, 1.0)
            rotated = cv2.warpAffine(query_img, M, (w, h))
            
            kp_query, desc_query = self.extract_sift_features(rotated)
            matches = self.match_sift_features(desc_template, desc_query)
            
            if len(matches) > best_score:
                best_score = len(matches)
                best_matches = matches
                best_angle = angle
        
        return best_matches, best_angle, best_score

    def load_category_templates(self, base_folder, force_recompute=False):
        """
        加载所有类别的模板图像
        
        参数:
            base_folder: 基础文件夹路径
            force_recompute: 是否强制重新计算特征（忽略缓存）
        """
        # 尝试从缓存加载
        if not force_recompute and self._load_features_cache(base_folder):
            return
        
        print(f"正在加载模板图像...")
        
        category_folders = [d for d in os.listdir(base_folder) 
                           if os.path.isdir(os.path.join(base_folder, d))]
        
        for category in category_folders:
            category_path = os.path.join(base_folder, category)
            image_files = [f for f in os.listdir(category_path) 
                          if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
            
            self.category_templates[category] = []
            
            for img_file in tqdm(image_files, desc=f"处理 {category}", leave=False):
                img_path = os.path.join(category_path, img_file)
                img = cv2.imread(img_path)
                
                if img is None:
                    continue
                
                # 提取特征
                kp_sift, desc_sift = self.extract_sift_features(img)
                deep_feat = self.extract_deep_features(img)
                
                template_info = {
                    'filename': img_file,
                    'path': img_path,
                    'kp_sift': kp_sift,
                    'desc_sift': desc_sift,
                    'deep_features': deep_feat
                }
                
                self.category_templates[category].append(template_info)
            
            print(f"  类别 '{category}': 加载 {len(image_files)} 张模板")
        
        print(f"总共加载 {len(category_folders)} 个类别")
        
        # 保存特征到缓存
        self._save_features_cache(base_folder)

    def classify_single_image(self, query_img, use_rotation_correction=True):
        """对单个查询图像进行分类"""
        kp_query, desc_query = self.extract_sift_features(query_img)
        deep_feat_query = self.extract_deep_features(query_img)
        
        category_scores = {}
        category_details = {}
        
        for category, templates in self.category_templates.items():
            sift_scores = []
            deep_scores = []
            rotation_angles = []
            
            for template in templates:
                if use_rotation_correction:
                    matches, angle, score = self.match_with_rotation(
                        cv2.imread(template['path']), query_img
                    )
                    rotation_angles.append(angle)
                else:
                    matches = self.match_sift_features(
                        template['desc_sift'], desc_query
                    )
                    score = len(matches)
                    rotation_angles.append(0)
                
                sift_scores.append(score)
                
                deep_sim = 1 - cosine(template['deep_features'], deep_feat_query)
                deep_scores.append(deep_sim)
            
            avg_sift_score = np.mean(sift_scores) if sift_scores else 0
            avg_deep_score = np.mean(deep_scores) if deep_scores else 0
            max_sift_score = np.max(sift_scores) if sift_scores else 0
            max_deep_score = np.max(deep_scores) if deep_scores else 0
            
            sift_confidence = min(1.0, max_sift_score / 15.0)
            deep_confidence = max_deep_score
            
            final_score = sift_confidence * 0.6 + deep_confidence * 0.4
            
            category_scores[category] = final_score
            category_details[category] = {
                'sift_scores': sift_scores,
                'deep_scores': deep_scores,
                'avg_sift_score': avg_sift_score,
                'avg_deep_score': avg_deep_score,
                'max_sift_score': max_sift_score,
                'max_deep_score': max_deep_score,
                'sift_confidence': sift_confidence,
                'deep_confidence': deep_confidence,
                'rotation_angles': rotation_angles
            }
        
        sorted_categories = sorted(category_scores.items(), 
                                  key=lambda x: x[1], reverse=True)
        
        best_category = sorted_categories[0][0]
        best_score = sorted_categories[0][1]
        
        result = {
            'predicted_category': best_category,
            'confidence': best_score,
            'all_scores': dict(sorted_categories),
            'details': category_details,
            'top_3': sorted_categories[:3]
        }
        
        return result

    def classify_batch(self, query_folder, output_json=None, output_csv=None):
        """批量分类查询图像"""
        query_files = [f for f in os.listdir(query_folder) 
                      if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
        
        results = []
        
        print(f"正在分类 {len(query_files)} 张查询图像...\n")
        
        for query_file in tqdm(query_files, desc="分类进度"):
            query_path = os.path.join(query_folder, query_file)
            query_img = cv2.imread(query_path)
            
            if query_img is None:
                continue
            
            result = self.classify_single_image(query_img)
            result['query_file'] = query_file
            results.append(result)
        
        if output_json:
            self._save_json_results(results, output_json)
        
        if output_csv:
            self._save_csv_results(results, output_csv)
        
        return results

    def _save_json_results(self, results, output_path):
        """保存为 JSON"""
        json_results = []
        
        for result in results:
            json_result = {
                'query_file': result['query_file'],
                'predicted_category': result['predicted_category'],
                'confidence': float(result['confidence']),
                'all_scores': {k: float(v) for k, v in result['all_scores'].items()},
                'top_3': [(cat, float(score)) for cat, score in result['top_3']]
            }
            json_results.append(json_result)
        
        with open(output_path, 'w', encoding='utf-8') as f:
            json.dump(json_results, f, ensure_ascii=False, indent=2)
        
        print(f"JSON 结果已保存到: {output_path}")

    def _save_csv_results(self, results, output_path):
        """保存为 CSV"""
        import csv
        
        with open(output_path, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(['查询文件', '预测类别', '置信度', '所有类别得分'])
            
            for result in results:
                scores_str = '; '.join([f"{k}:{v:.4f}" 
                                       for k, v in result['all_scores'].items()])
                writer.writerow([
                    result['query_file'],
                    result['predicted_category'],
                    f"{result['confidence']:.4f}",
                    scores_str
                ])
        
        print(f"CSV 结果已保存到: {output_path}")

    def print_classification_report(self, result):
        """打印分类报告"""
        print("\n" + "="*70)
        print("分类结果")
        print("="*70)
        print(f"查询文件: {result['query_file']}")
        print(f"预测类别: {result['predicted_category']}")
        print(f"置信度: {result['confidence']:.4f}")
        print("\n所有类别得分:")
        
        for category, score in result['all_scores'].items():
            print(f"  {category}: {score:.4f}")
        
        print("\nTop 3 预测:")
        for i, (category, score) in enumerate(result['top_3'], 1):
            details = result['details'][category]
            print(f"  {i}. {category} (置信度: {score:.4f})")
            print(f"     - SIFT 最高分: {details['max_sift_score']:.1f}")
            print(f"     - 深度学习最高分: {details['max_deep_score']:.4f}")
            print(f"     - 旋转角度: {details['rotation_angles']}")
            print("="*70 + "\n")


# ==================== 使用示例 ====================
if __name__ == "__main__":

    # 初始化分类器（指定缓存目录）
    classifier = PharmaceuticalBottleClassifier(cache_dir='./feature_cache')

    # 加载模板（多个类别文件夹）
    base_template_folder = r"A:\\Temp_dataset\\Blured"
    
    # 首次运行：force_recompute=False（默认），会自动保存特征到缓存
    # 后续运行：直接从缓存加载，速度快很多
    classifier.load_category_templates(base_template_folder, force_recompute=False)
    
    # 如果需要重新计算特征（比如模板更新了），使用：
    # classifier.load_category_templates(base_template_folder, force_recompute=True)

    # ========== 方式1: 单个图像分类 ==========
    print("\n方式1: 单个图像分类")
    print("-" * 70)

    query_image_path = "./YaoPing_test_true/Cropped_Resized/frame_000293_2.png"
    query_img = cv2.imread(query_image_path)

    result = classifier.classify_single_image(query_img, use_rotation_correction=True)
    result['query_file'] = os.path.basename(query_image_path)

    classifier.print_classification_report(result)

    # ========== 方式2: 批量分类 ==========
    # print("\n方式2: 批量分类")
    # print("-" * 70)

    # query_folder = r"A:\项目\医药物品检测\DataBase_Data\YaoPing\YaoPing_test_true"

    # batch_results = classifier.classify_batch(
    #     query_folder,
    #     output_json=r"A:\项目\医药物品检测\classification_results.json",
    #     output_csv=r"A:\项目\医药物品检测\classification_results.csv"
    # )

    # # 打印前几个结果
    # for result in batch_results[:3]:
    #     classifier.print_classification_report(result)

    # # 统计准确率（如果有真实标签）
    # print("\n分类统计:")
    # print("-" * 70)
    # category_counts = {}
    # for result in batch_results:
    #     cat = result['predicted_category']
    #     category_counts[cat] = category_counts.get(cat, 0) + 1

    # for category, count in sorted(category_counts.items()):
    #     print(f"  {category}: {count} 张")
```

# 均值特征识别

```python
import cv2
import numpy as np
import torch
import torchvision.models as models
from torchvision import transforms
from PIL import Image
import os
from pathlib import Path
from scipy.spatial.distance import cosine
import json
from tqdm import tqdm
import pickle

class PharmaceuticalBottleClassifier:
    """药瓶多类别分类器 - 优化版本"""

    def __init__(self, device='cuda', cache_dir='./feature_cache'):
        self.device = torch.device(device if torch.cuda.is_available() else 'cpu')
        self.cache_dir = cache_dir
        
        os.makedirs(cache_dir, exist_ok=True)
        
        # SIFT 特征提取
        self.sift = cv2.SIFT_create()
        
        # 深度学习特征提取 - 使用更轻量的模型
        self.model = models.mobilenet_v3_small(pretrained=True)
        self.model = torch.nn.Sequential(*list(self.model.children())[:-1])
        self.model.eval()
        self.model = self.model.to(self.device)
        
        self.preprocess = transforms.Compose([
            transforms.Resize(224),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                               std=[0.229, 0.224, 0.225]),
        ])
        
        # 特征匹配器
        FLANN_INDEX_KDTREE = 1
        index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
        search_params = dict(checks=50)
        self.flann = cv2.FlannBasedMatcher(index_params, search_params)
        
        self.category_templates = {}
        self.category_deep_features = {}  # 缓存深度特征均值

    def _get_cache_path(self, base_folder):
        """获取缓存文件路径"""
        folder_hash = str(hash(os.path.abspath(base_folder)))
        cache_file = os.path.join(self.cache_dir, f'features_{folder_hash}.pkl')
        return cache_file

    def _save_features_cache(self, base_folder):
        """保存特征到本地缓存"""
        cache_path = self._get_cache_path(base_folder)
        
        cache_data = {}
        
        for category, templates in self.category_templates.items():
            cache_data[category] = []
            
            for template in templates:
                kp_data = []
                if template['kp_sift']:
                    for kp in template['kp_sift']:
                        kp_data.append({
                            'pt': kp.pt,
                            'size': kp.size,
                            'angle': kp.angle,
                            'response': kp.response,
                            'octave': kp.octave,
                            'class_id': kp.class_id
                        })
                
                cache_item = {
                    'filename': template['filename'],
                    'path': template['path'],
                    'kp_sift': kp_data,
                    'desc_sift': template['desc_sift'],
                    'deep_features': template['deep_features']
                }
                cache_data[category].append(cache_item)
        
        with open(cache_path, 'wb') as f:
            pickle.dump(cache_data, f)
        
        print(f"特征已保存到: {cache_path}\n")

    def _load_features_cache(self, base_folder):
        """从缓存加载特征"""
        cache_path = self._get_cache_path(base_folder)
        
        if not os.path.exists(cache_path):
            return False
        
        try:
            with open(cache_path, 'rb') as f:
                cache_data = pickle.load(f)
            
            for category, templates in cache_data.items():
                self.category_templates[category] = []
                
                for template in templates:
                    kp_list = []
                    for kp_data in template['kp_sift']:
                        kp = cv2.KeyPoint(
                            x=kp_data['pt'][0],
                            y=kp_data['pt'][1],
                            size=kp_data['size'],
                            angle=kp_data['angle'],
                            response=kp_data['response'],
                            octave=kp_data['octave'],
                            class_id=kp_data['class_id']
                        )
                        kp_list.append(kp)
                    
                    template_info = {
                        'filename': template['filename'],
                        'path': template['path'],
                        'kp_sift': kp_list,
                        'desc_sift': template['desc_sift'],
                        'deep_features': template['deep_features']
                    }
                    self.category_templates[category].append(template_info)
            
            # 计算每个类别的深度特征均值
            self._compute_category_deep_features()
            
            print(f"从缓存加载特征: {cache_path}")
            print(f"加载了 {len(self.category_templates)} 个类别\n")
            return True
        
        except Exception as e:
            print(f"加载缓存失败: {e}")
            return False

    def _compute_category_deep_features(self):
        """计算每个类别的深度特征均值"""
        for category, templates in self.category_templates.items():
            if templates:
                deep_feats = np.array([t['deep_features'] for t in templates])
                self.category_deep_features[category] = np.mean(deep_feats, axis=0)

    def preprocess_image(self, image):
        """预处理图像"""
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        enhanced = clahe.apply(gray)
        return enhanced

    def extract_sift_features(self, image):
        """提取 SIFT 特征"""
        processed = self.preprocess_image(image)
        kp, desc = self.sift.detectAndCompute(processed, None)
        return kp, desc

    def extract_deep_features(self, image):
        """提取深度学习特征"""
        img_pil = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        img_tensor = self.preprocess(img_pil).unsqueeze(0).to(self.device)
        
        with torch.no_grad():
            features = self.model(img_tensor)
            features = features.squeeze().cpu().numpy()
        
        return features

    def match_sift_features(self, desc1, desc2):
        """SIFT 特征匹配 - 优化版本"""
        if desc1 is None or desc2 is None:
            return 0
        
        try:
            matches = self.flann.knnMatch(desc1, desc2, k=2)
            good_count = 0
            
            for match_pair in matches:
                if len(match_pair) == 2:
                    m, n = match_pair
                    if m.distance < 0.7 * n.distance:
                        good_count += 1
            
            return good_count
        except:
            return 0

    def load_category_templates(self, base_folder, force_recompute=False):
        """加载所有类别的模板图像"""
        if not force_recompute and self._load_features_cache(base_folder):
            return
        
        print(f"正在加载模板图像...")
        
        category_folders = [d for d in os.listdir(base_folder) 
                           if os.path.isdir(os.path.join(base_folder, d))]
        
        for category in category_folders:
            category_path = os.path.join(base_folder, category)
            image_files = [f for f in os.listdir(category_path) 
                          if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
            
            self.category_templates[category] = []
            
            for img_file in tqdm(image_files, desc=f"处理 {category}", leave=False):
                img_path = os.path.join(category_path, img_file)
                img = cv2.imread(img_path)
                
                if img is None:
                    continue
                
                kp_sift, desc_sift = self.extract_sift_features(img)
                deep_feat = self.extract_deep_features(img)
                
                template_info = {
                    'filename': img_file,
                    'path': img_path,
                    'kp_sift': kp_sift,
                    'desc_sift': desc_sift,
                    'deep_features': deep_feat
                }
                
                self.category_templates[category].append(template_info)
            
            print(f"  类别 '{category}': 加载 {len(image_files)} 张模板")
        
        print(f"总共加载 {len(category_folders)} 个类别")
        
        # 计算类别深度特征均值
        self._compute_category_deep_features()
        
        # 保存特征到缓存
        self._save_features_cache(base_folder)

    def classify_single_image(self, query_img):
        """对单个查询图像进行分类 - 优化版本"""
        kp_query, desc_query = self.extract_sift_features(query_img)
        deep_feat_query = self.extract_deep_features(query_img)
        
        category_scores = {}
        category_details = {}
        
        for category, templates in self.category_templates.items():
            sift_scores = []
            deep_scores = []
            
            # SIFT 匹配
            for template in templates:
                score = self.match_sift_features(template['desc_sift'], desc_query)
                sift_scores.append(score)
            
            # 深度学习特征相似度 - 与类别均值比较
            category_mean_feat = self.category_deep_features[category]
            deep_sim = 1 - cosine(category_mean_feat, deep_feat_query)
            
            max_sift_score = np.max(sift_scores) if sift_scores else 0
            
            # 归一化
            sift_confidence = min(1.0, max_sift_score / 15.0)
            deep_confidence = max(0, deep_sim)
            
            # 加权融合
            final_score = sift_confidence * 0.6 + deep_confidence * 0.4
            
            category_scores[category] = final_score
            category_details[category] = {
                'max_sift_score': max_sift_score,
                'deep_confidence': deep_confidence,
                'sift_confidence': sift_confidence
            }
        
        sorted_categories = sorted(category_scores.items(), 
                                  key=lambda x: x[1], reverse=True)
        
        best_category = sorted_categories[0][0]
        best_score = sorted_categories[0][1]
        
        result = {
            'predicted_category': best_category,
            'confidence': best_score,
            'all_scores': dict(sorted_categories),
            'details': category_details,
            'top_3': sorted_categories[:3]
        }
        
        return result

    def classify_batch(self, query_folder, output_json=None, output_csv=None):
        """批量分类查询图像"""
        query_files = [f for f in os.listdir(query_folder) 
                      if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
        
        results = []
        
        print(f"正在分类 {len(query_files)} 张查询图像...\n")
        
        for query_file in tqdm(query_files, desc="分类进度"):
            query_path = os.path.join(query_folder, query_file)
            query_img = cv2.imread(query_path)
            query_img = cv2.resize(query_img,(224,224))
            if query_img is None:
                continue
            
            result = self.classify_single_image(query_img)
            result['query_file'] = query_file
            results.append(result)
        print(results)
        if output_json:
            self._save_json_results(results, output_json)
        
        if output_csv:
            self._save_csv_results(results, output_csv)
        
        return results

    def _save_json_results(self, results, output_path):
        """保存为 JSON"""
        json_results = []
        
        for result in results:
            json_result = {
                'query_file': result['query_file'],
                'predicted_category': result['predicted_category'],
                'confidence': float(result['confidence']),
                'all_scores': {k: float(v) for k, v in result['all_scores'].items()},
                'top_3': [(cat, float(score)) for cat, score in result['top_3']]
            }
            json_results.append(json_result)
        
        with open(output_path, 'w', encoding='utf-8') as f:
            json.dump(json_results, f, ensure_ascii=False, indent=2)
        
        print(f"JSON 结果已保存到: {output_path}")

    def _save_csv_results(self, results, output_path):
        """保存为 CSV"""
        import csv
        
        with open(output_path, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(['查询文件', '预测类别', '置信度', '所有类别得分'])
            
            for result in results:
                scores_str = '; '.join([f"{k}:{v:.4f}" 
                                       for k, v in result['all_scores'].items()])
                writer.writerow([
                    result['query_file'],
                    result['predicted_category'],
                    f"{result['confidence']:.4f}",
                    scores_str
                ])
        
        print(f"CSV 结果已保存到: {output_path}")

    def print_classification_report(self, result):
        """打印分类报告"""
        print("\n" + "="*70)
        print("分类结果")
        print("="*70)
        print(f"查询文件: {result['query_file']}")
        print(f"预测类别: {result['predicted_category']}")
        print(f"置信度: {result['confidence']:.4f}")
        print("\n所有类别得分:")
        
        for category, score in result['all_scores'].items():
            print(f"  {category}: {score:.4f}")
        
        print("\nTop 3 预测:")
        for i, (category, score) in enumerate(result['top_3'], 1):
            details = result['details'][category]
            print(f"  {i}. {category} (置信度: {score:.4f})")
            print(f"     - SIFT 最高分: {details['max_sift_score']:.1f}")
            print(f"     - 深度学习得分: {details['deep_confidence']:.4f}")
        
        print("="*70 + "\n")


# ==================== 使用示例 ====================
if __name__ == "__main__":

    # 初始化分类器
    classifier = PharmaceuticalBottleClassifier(cache_dir='./feature_cache')

    # 加载模板
    base_template_folder = r"A:\\Temp_dataset\\Blured"
    classifier.load_category_templates(base_template_folder, force_recompute=False)

    # ========== 方式1: 单个图像分类 ==========
    print("\n方式1: 单个图像分类")
    print("-" * 70)

    query_image_path = "./YaoPing_test_true/Cropped_Resized/frame_000293_2.png"
    query_img = cv2.imread(query_image_path)
    query_img = cv2.resize(query_img,(224,224))
    result = classifier.classify_single_image(query_img)
    result['query_file'] = os.path.basename(query_image_path)

    classifier.print_classification_report(result)

    # ========== 方式2: 批量分类 ==========
    print("\n方式2: 批量分类")
    print("-" * 70)

    query_folder = "A:\Temp_dataset\Test_0312"

    batch_results = classifier.classify_batch(
        query_folder,
        output_json="A:\\项目\\医药物品检测\\classification_results.json",
        output_csv="A:\\项目\\医药物品检测\\classification_results.csv"
    )

    for result in batch_results[:3]:
        classifier.print_classification_report(result)

    print("\n分类统计:")
    print("-" * 70)
    category_counts = {}
    for result in batch_results:
        cat = result['predicted_category']
        category_counts[cat] = category_counts.get(cat, 0) + 1

    for category, count in sorted(category_counts.items()):
        print(f"  {category}: {count} 张")
```

## 单图像OCR识别

```python
import easyocr
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os

class EasyOCRProcessor:
    def __init__(self, languages=['ch_sim', 'en']):
        """
        初始化EasyOCR读取器
        
        Args:
            languages: 支持的语言列表，默认支持简体中文和英文
        """
        print("正在初始化EasyOCR读取器...")
        self.reader = easyocr.Reader(languages, gpu=True)  # 如果有GPU则使用GPU加速
        print("EasyOCR读取器初始化完成")

    def read_image(self, image_path):
        """
        读取并识别图像中的文字
        
        Args:
            image_path: 图像文件路径
            
        Returns:
            results: 识别结果列表
        """
        if not os.path.exists(image_path):
            raise FileNotFoundError(f"图像文件不存在: {image_path}")
        
        print(f"正在处理图像: {image_path}")
        
        # 使用EasyOCR进行文字识别
        results = self.reader.readtext(image_path)
        
        return results

    def process_results(self, results):
        """
        处理识别结果
        
        Args:
            results: EasyOCR识别结果
            
        Returns:
            processed_data: 处理后的数据
        """
        processed_data = []
        
        for (bbox, text, confidence) in results:
            # 提取边界框坐标
            top_left = tuple(map(int, bbox[0]))
            top_right = tuple(map(int, bbox[1]))
            bottom_right = tuple(map(int, bbox[2]))
            bottom_left = tuple(map(int, bbox[3]))
            
            processed_data.append({
                'text': text,
                'confidence': confidence,
                'bbox': {
                    'top_left': top_left,
                    'top_right': top_right,
                    'bottom_right': bottom_right,
                    'bottom_left': bottom_left
                }
            })
        
        return processed_data

    def draw_results(self, image_path, results, output_path=None):
        """
        在图像上绘制识别结果
        
        Args:
            image_path: 原始图像路径
            results: 识别结果
            output_path: 输出图像路径（可选）
        """
        # 读取图像
        image = cv2.imread(image_path)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        # 在图像上绘制结果
        for (bbox, text, confidence) in results:
            # 绘制边界框
            points = np.array(bbox, dtype=np.int32)
            cv2.polylines(image_rgb, [points], True, (0, 255, 0), 2)
            
            # 添加文字标签
            top_left = tuple(map(int, bbox[0]))
            cv2.putText(image_rgb, f'{text} ({confidence:.2f})', 
                       (top_left[0], top_left[1] - 10),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
        
        # 显示结果
        plt.figure(figsize=(12, 8))
        plt.imshow(image_rgb)
        plt.axis('off')
        plt.title('EasyOCR识别结果')
        plt.tight_layout()
        
        if output_path:
            plt.savefig(output_path, dpi=300, bbox_inches='tight')
            print(f"结果图像已保存到: {output_path}")
        
        plt.show()

    def extract_text_only(self, results, min_confidence=0.5):
        """
        仅提取文字内容
        
        Args:
            results: 识别结果
            min_confidence: 最小置信度阈值
            
        Returns:
            text_list: 文字列表
        """
        text_list = []
        for (bbox, text, confidence) in results:
            if confidence >= min_confidence:
                text_list.append(text)
        
        return text_list

    def save_results_to_file(self, results, output_file):
        """
        将识别结果保存到文件
        
        Args:
            results: 识别结果
            output_file: 输出文件路径
        """
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write("EasyOCR识别结果\n")
            f.write("=" * 50 + "\n\n")
            
            for i, (bbox, text, confidence) in enumerate(results, 1):
                f.write(f"文本块 {i}:\n")
                f.write(f"内容: {text}\n")
                f.write(f"置信度: {confidence:.4f}\n")
                f.write(f"边界框: {bbox}\n")
                f.write("-" * 30 + "\n")
        
        print(f"识别结果已保存到: {output_file}")


def main():
    """
    主函数 - 使用示例
    """
    # 初始化OCR处理器
    ocr_processor = EasyOCRProcessor(languages=['ch_sim', 'en'])
    
    # 图像文件路径（请替换为你的图像路径）
    image_path = "A:\项目\医药物品检测\DataBase_Data\YaoPing_test_true\Cropped\\YaoDai_Test1.png"  # 替换为实际的图像路径
    
    try:
        # 进行OCR识别
        results = ocr_processor.read_image(image_path)
        
        # 打印原始结果
        print("\n=== 原始识别结果 ===")
        for i, (bbox, text, confidence) in enumerate(results, 1):
            print(f"文本块 {i}: '{text}' (置信度: {confidence:.4f})")
        
        # 处理结果
        processed_data = ocr_processor.process_results(results)
        
        # 打印处理后的结果
        print("\n=== 处理后的结果 ===")
        for i, data in enumerate(processed_data, 1):
            print(f"文本块 {i}:")
            print(f"  内容: {data['text']}")
            print(f"  置信度: {data['confidence']:.4f}")
            print(f"  边界框: {data['bbox']}")
        
        # 提取纯文字内容
        text_only = ocr_processor.extract_text_only(results, min_confidence=0.5)
        print("\n=== 提取的文字内容 ===")
        for text in text_only:
            print(f"- {text}")
        
        # 绘制结果（如果图像存在）
        if os.path.exists(image_path):
            ocr_processor.draw_results(image_path, results, "ocr_result.png")
        
        # 保存结果到文件
        ocr_processor.save_results_to_file(results, "ocr_results.txt")
        
    except FileNotFoundError as e:
        print(f"错误: {e}")
        print("请确保图像文件存在，或者创建一个测试图像")
        
        # 创建一个简单的测试图像
        create_test_image()
        
    except Exception as e:
        print(f"处理过程中出现错误: {e}")


def create_test_image():
    """
    创建一个包含文字的测试图像
    """
    from PIL import Image, ImageDraw, ImageFont
    
    # 创建白色背景图像
    img = Image.new('RGB', (800, 400), color='white')
    draw = ImageDraw.Draw(img)
    
    # 添加一些测试文字
    try:
        # 尝试使用系统字体
        font = ImageFont.truetype("arial.ttf", 40)
    except:
        # 如果没有找到字体，使用默认字体
        font = ImageFont.load_default()
    
    # 绘制文字
    draw.text((50, 50), "Hello World!", fill='black', font=font)
    draw.text((50, 120), "EasyOCR测试", fill='blue', font=font)
    draw.text((50, 190), "OCR Recognition", fill='red', font=font)
    draw.text((50, 260), "图像文字识别", fill='green', font=font)
    
    # 保存测试图像
    img.save("test_image.jpg")
    print("已创建测试图像: test_image.jpg")


# 简化版本的使用示例
def simple_ocr_example(image_path):
    """
    简化版OCR识别示例
    
    Args:
        image_path: 图像文件路径
    """
    # 初始化EasyOCR读取器
    reader = easyocr.Reader(['ch_sim', 'en'])
    
    # 读取图像并进行OCR识别
    results = reader.readtext(image_path)
    
    # 打印结果
    print("识别结果:")
    for (bbox, text, confidence) in results:
        print(f"文字: {text}")
        print(f"置信度: {confidence:.4f}")
        print(f"位置: {bbox}")
        print("-" * 30)
    
    return results


if __name__ == "__main__":
    main()
```
